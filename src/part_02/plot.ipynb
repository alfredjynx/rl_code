```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/alpha0.1-gamma0.6.csv')
plt.plot(df['Episodes'],df['Actions'])
plt.xlabel('Episodes')
plt.ylabel('# Actions')
plt.title('# Actions vs Episodes')
plt.show()
```

```python
df = pd.read_csv('data/alpha0.01-gamma0.6.csv')
plt.plot(df['Episodes'],df['Actions'])
plt.xlabel('Episodes')
plt.ylabel('# Actions')
plt.title('# Actions vs Episodes')
plt.show()
```

```python
df = pd.read_csv('data/alpha0.001-gamma0.6.csv')
plt.plot(df['Episodes'],df['Actions'])
plt.xlabel('Episodes')
plt.ylabel('# Actions')
plt.title('# Actions vs Episodes')
plt.show()
```

```python
df01 = pd.read_csv('data/alpha0.1-gamma0.6.csv')
df001 = pd.read_csv('data/alpha0.01-gamma0.6.csv')
df0001 = pd.read_csv('data/alpha0.001-gamma0.6.csv')
plt.plot(df0001['Episodes'],df0001['Actions'], label='0.001', alpha=0.7)
plt.plot(df001['Episodes'],df001['Actions'], label='0.01', alpha=0.7)
plt.plot(df01['Episodes'],df01['Actions'], label='0.1', alpha=0.7)
plt.xlabel('Episodes')
plt.ylabel('# Actions')
plt.title('# Actions vs Episodes')
plt.legend(title='Alpha')
plt.savefig('results/actions_X_episodes_alphas')
plt.show()
```

A figura ilustra um fenmeno que ocorre quando diminuimos o alpha: o grfico demora mais tempo para se estabilizar e convergir ao ponto desejado. Isso ocorre graas ao fato do alpha ser o coeficiente de aprendizado, ou seja, o quanto o algoritmo valoriza a reward do estado aprendido, assim, quando menor for esse valor menos ele ir aprender por iterao, levando mais tempo para realmente convergir ao ponto que gostaramos (muitas vezes nem convergindo).
